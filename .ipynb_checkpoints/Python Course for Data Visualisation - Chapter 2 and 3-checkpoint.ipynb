## Filtering

```python
#Filter data to have car type 'Toyota Sienna' and an interest rate of 0.0702
df.head()
#check to see the number of different items and their counts
```

```python
df['car_type'].value_counts()
```

```python
car_filter = df['car_type']=='Toyota Sienna'
car_filter.head()
#returns true or false on whether each row has a toyota sienna
```

```python
#two approaches to using this filter
# use square brackets to make DF
df[car_filter].head()
```

```python
#or use loc
df.loc[car_filter, :] #often this approach is more legible
```

```python
#make sure to assign this filter to a new dataframe or overwrite old one
df = df.loc[car_filter, :] 
df['car_type'].value_counts()
```

```python
#now you need to the same for the interest rate
df['interest_rate'].value_counts()
```

```python
interestfilt = df['interest_rate'] == 0.0702
interestfilt
```

```python
df = df.loc[interestfilt,:]
df['interest_rate'].value_counts()
df
```

```python
#how to filter both at the same time
df.loc[car_filter & interestfilt, :]
```

## Removing or filling in Missing Data

```python
#don't want to graph missing data and therefore draw wronf conclusions
df.info()
```

```python
#remove missing data with dropna()
df[30:40].dropna(how = 'any') #knowing where the null value is and dropping it - how variable means drop the entire index row
```

```python
df['Interest Paid'][30:40].fillna(0) #filling with a zero - but the NaN could mean something so this often is that good at it
```

```python
df['Interest Paid'][30:40].fillna(method = 'bfill') #index after is 89.77 done with time series data - backfill
```

```python
df['Interest Paid'][30:40].fillna(method = 'ffill') #forward fill - takes value before missing value and puts it where the NaN. 
#which one to do depends on application
```

```python
df['Interest Paid'][30:40].interpolate(method = 'linear') # you can also inetrpolate - uses a linear model to fill in missing value
#93 is between 89 and 96
```

```python
#sum method by defualt ignores nan so this can change the value
#try and get the most accurate values
```
