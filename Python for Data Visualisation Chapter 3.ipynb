## Marker Type and Colour

```python
#Markers can display more info to the users
plt.style.use('seaborn')
plt.plot(month_number, interest_paid, marker = '.', markersize=10)
plt.plot(month_number, principal_paid, marker = '.', markersize=10)
#similar to whats been plotted
```

```python
#can also change the colour using c paramter
plt.style.use('seaborn')
plt.plot(month_number, interest_paid, c= 'b', marker = '.', markersize=10)
plt.plot(month_number, principal_paid, c='black', marker = '.', markersize=10)
#similar to whats been plotted
```

```python
#can accept hex colours, also RGB tuples
```

## Titles Labels and Limits

```python
#Matlab style, x lim and y lim
plt.style.use('seaborn')
plt.plot(month_number, interest_paid)
plt.plot(month_number, principal_paid)
plt.xlim(left=1, right =70)
plt.ylim(bottom = 0, top=1000)
plt.xlabel('Month', fontsize = 20)
plt.ylabel('Dollars', fontsize = 20)
plt.title('Interest and Principal Paid per Month', fontsize = 30)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
```

```python
fig, axes = plt.subplots(nrows=1, ncols = 1)
axes.plot(month_number, interest_paid, c = 'k') #k = black
axes.plot(month_number, principal_paid, c = 'b') 
axes.set_xlabel('Month', fontsize = 20)
axes.set_ylabel('Dollars', fontsize = 20)
axes.set_title('Interest and Principal Paid per month')
#very similar to the matlab style just different functions
```

## Grid Lines

```python
#Can be appealing but also can help identify fine differences between values
plt.style.use('ggplot')
plt.plot(month_number, interest_paid, label = 'Interest')
plt.plot(month_number, principal_paid, label = 'Principal')
plt.grid(axis = 'y', c = 'k')
plt.legend(loc = 'center right') # make sure you have labels in the data, loc paramater helps you chnage where it is located
#can also do coordinates
#specify which way you want vertical gridlines, colour and style etc
#legends help define plot elements but you dont want to cover up any data
```
